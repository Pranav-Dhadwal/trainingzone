#### Question 1 :  What is the difference between a List and a Dictionary in Python?

---

while list is just a simple collection that can also replicated behaviour of 1d 2d 3d and so on

Use a list when order matters and you're storing a sequence of similar items — like a list of scores, names, or steps. Use a dict when you need to label your data or look something up by name

```python
# question 2 ----------

def calculate(numbers):
    total = 0
    for num in numbers:
        total += num
    return total / len(numbers)

result = calculate([10, 20, 30, 40])
print(result)

# Two parts:

# What does this function do?
# What happens if you pass an empty list [] — and how would you fix it? 🎯
```

```python
# result -> calculate[list] ->  total -> loop -> condition -> return -> print 

def calculate(numbers):
    if not numbers:
        return "Cannot calculate average of an empty list"
    return sum(numbers) / len(numbers)
```

#### Question 3 : Write a function that takes a list of numbers and returns only the even numbers using list comprehension.
Then tell me — what would change if you used a regular for loop instead? 🎯

```python
def even (nums):
    if not nums:
        return "List is empty"
    return [i for i in nums if i%2 == 0]

result = even([1,2,3])
print("No even present" if not result else result)

# For loop — builds list incrementally, calls .append() each iteration
# List comprehension — pre-allocates memory, runs at C speed
# For large datasets this difference is actually measurable
```
