```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'name' : ['bunty' ,'chunty'],
    'email' : ['for@go', 'hah@sad'],
    'gender' : ['M', 'F'],
    'age' : [1,256],
    'date' : [5,9]
})

df.duplicated().sum()     # count duplicates
df.drop_duplicates()      # remove them
df.drop_duplicates(subset=['name', 'email'])  # based on specific columns

```

```python
# fixing types 
df.dtypes                        # check types
df['age'] = df['age'].astype(int)        # convert to int
df['date'] = pd.to_datetime(df['date'])  # convert to datetime



# note 
# This will crash if any value is '25 years' or '' or NaN
df['age'] = df['age'].astype(int)

# Safer approach for real world data
df['age'] = pd.to_numeric(df['age'], errors='coerce')
# errors='coerce' → converts bad values to NaN instead of crashing
```

```python
# string cleaning
df['name'].str.lower()           # lowercase
df['name'].str.strip()           # remove whitespace
df['name'].str.replace('_', ' ') # replace characters
df['name'].str.contains('Alice') # filter by string
```

```python
# replacing wrong values 
df['gender'] = df['gender'].replace({'M': 'Male', 'F': 'Female'})
df['age'].replace(-1, np.nan)  # replace impossible values with NaN
```

```python
# renaming columns 

df.rename(columns={'name': 'N', 'gender': 'g'}, inplace=True)
df.columns = df.columns.str.lower().str.strip()  # clean all at once
```

```python
df['age'].clip(lower=0, upper=30)  # cap values between 0 and 300
```

```python
df = pd.DataFrame({
    'name': ['  Alice', 'BOB  ', '  charlie  ', 'DAVID'],
    'city': ['new york', 'NEW YORK', 'New York', 'new york']
})
```

```python
df['name'] = df['name'].str.strip().str.title()
df['city'] = df['city'].str.lower()
```

```python

df = pd.DataFrame({
    'id':     [1, 2, 3, 2, 4],
    'gender': ['M', 'F', 'male', 'F', 'FEMALE'],
    'age':    [25, 30, -1, 30, 28]
})
```

```python
df
```

```python
# Standardize the gender column so all values are either 'Male' or 'Female'
df['gender'] = df['gender'].str.lower().replace({
    'm': 'Male', 
    'f': 'Female',
    'male': 'Male',
    'female': 'Female'
})
# Replace -1 in age with NaN
df['age'] = df['age'].replace(-1, np.nan)

# Remove duplicate rows based on the id column
df = df.drop_duplicates(subset='id')
```

```python
# optimised 

df['gender'] = df['gender'].str.lower().str[0].replace({
    'm': 'Male', 
    'f': 'Female'
})
# str[0] grabs first letter → 'm' or 'f' always → clean mapping
```

```python
df = pd.DataFrame({
    'age':    [25, 30, 150, -5, 28],
    'salary': [50000, 60000, 10000000, 45000, 55000]
})
```

```python
# Clip age between 0 and 100
df['age'] = df['age'].clip(lower=0, upper=100)
# Clip salary between 20000 and 200000
df['salary'] = df['salary'].clip(lower=20000, upper=200000)
df
```
