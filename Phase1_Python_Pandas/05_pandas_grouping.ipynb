#### groupby()
- groups the data based on columns and allow us to perform operation on it 
- like : data.groupby('Class')['Score'].mean()


#### agg()
- allow us to perform more that one aggregation operation in a single time 
- like : data.groupby('Class')['Score].agg(['mean', 'max', 'min', 'count])
- second : data.grouby('Class').agg({ 'Score' : ['mean', 'max'], 'Age' : ['mean', 'max]})

#### pivot_table()
- its like a increament of gropby() with ability of grouping in 2 dimensions
- like - pd.pivot_table(df, 
values = 'Survived'
index = 'Pclass'
columns = 'gender'
aggfunc = 'mean'
)

```python
import pandas as pd
```

```python
data = pd.read_csv(r'../datasets/data_files/Titanic-Dataset.csv')
```

```python
data.info()
```

```python
# Question 1 > 
# Write code to find the average age and total fare paid for each passenger class (Pclass) — both in a single operation. 
```

```python
data.groupby('Pclass').agg({
    'Age' : 'mean',
    'Fare' : 'sum'
})
```

```python
# Question 2
# Using the same Titanic dataset — write a pivot table that shows the survival rate for each combination of Pclass and Sex. 🎯
```

```python
pd.pivot_table(data,
               values='Survived',
               index='Pclass', 
               columns='Sex',
               aggfunc='mean')
```

```python
# One query — three things:
# Using Titanic, write code to find for each Pclass:

# Average age
# Survival rate
# Total number of passengers

# All in a single agg() call. 🎯
```

```python
data.groupby('Pclass').agg({
    'Age' : 'mean',
    'Survived' : 'mean',
    'PassengerId' : 'count'
})
```

```python
# improvement 

data.groupby('Pclass').agg(
    avg_age        = ('Age', 'mean'),
    survival_rate  = ('Survived', 'mean'),
    total_passengers = ('PassengerId', 'count')
)
```

```python
df = pd.DataFrame({
    'city':     ['Delhi', 'Mumbai', 'Delhi', 'Mumbai', 'Delhi'],
    'category': ['A', 'B', 'A', 'A', 'B'],
    'sales':    [100, 200, 150, 300, 120],
    'profit':   [10, 40, 30, 60, 20]
})
df
```

```python
# Write a pivot table showing total sales for each city (rows) and category (columns).
# And tell me — what happens to missing combinations if a city has no sales in a certain category? 🎯
```

```python
pd.pivot_table(df,
               values='sales',
               index='city',
               columns='category',
               aggfunc='sum'
               )

# misssing combination may be filled with 0, but nooooo
```

```python
# correction 
# missing values are filled with NaN

# to handle it 

pd.pivot_table(df, values='sales', index='city',
               columns='category', aggfunc='sum',
               fill_value=0)  # ← fills NaN with 0
```

```python
df = pd.DataFrame({
    'department': ['HR', 'Tech', 'HR', 'Tech', 'HR', 'Tech'],
    'gender':     ['M', 'M', 'F', 'F', 'M', 'F'],
    'salary':     [50000, 90000, 45000, 95000, 52000, 88000],
    'experience': [2, 5, 3, 7, 4, 6]
})
df
```

```python
# Write a single query to find for each department:

# Average salary for Male and Female separately
# Maximum experience
```

```python
df.groupby(['department', 'gender']).agg(
    Average_Salary=('salary', 'mean'),
    Max_Experience=('experience', 'max')
)   
```
