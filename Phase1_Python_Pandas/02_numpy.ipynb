```python
scores = np.array([[85, 42, 91],
                   [78, 95, 33],
                   [60, 71, 88]])
```

```python
# Four parts:

# Get the mean score of each student (row-wise)
for i in range(0,3):
    print(scores[i].mean())

print(scores.mean(axis=1))
print(scores.mean(axis=0))

# Get the highest score in the entire array
print(scores.max())

# Get all scores above 70 using boolean indexing
scores[scores>70]

# What is the shape, size and ndim of this array? 
# shape -> 3,3
# size -> 9
# ndim -> 2
```

### axis=1 → traverse row by row → one result per row
### axis=0 → traverse col by col → one result per col

#### Takeaways 

- ndim — number of dimensions
- axis=0 vs axis=1 — collapse rows vs columns
- dtype hierarchy — bool → int → float → complex → string
- Always prefer vectorized operations over loops in NumPy
