```python
data['Age'] = data.groupby('Pclass')['Age'].transform(
    lambda x: x.fillna(x.median())
)
```

```python
data = data.drop(columns='Cabin')
```

```python
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
```
